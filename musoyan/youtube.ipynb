```python
from pathlib import Path
import yt_dlp
import pandas as pd


def playlist_to_df(playlist_url: str) -> pd.DataFrame:
    """
    Return a DataFrame for every public video in a YouTube playlist with:
      • URL
      • Title
      • Duration (minutes, int)
      • Views
      • Number of likes        (None if hidden)
      • Number of dislikes     (requires ReturnYouTubeDislike; None if unavailable)
      • Number of comments     (None if disabled/hidden)
      • Playlist name
    No YouTube Data API key needed.
    """
    # Ask yt-dlp for full metadata on each video
    ydl_opts = {
        "quiet": True,
        "skip_download": True,
        "extract_flat": False,    # full per-video info → stats
        "dump_single_json": True,
    }

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        pl_info = ydl.extract_info(playlist_url, download=False)
    print(f"Extracted {len(pl_info['entries'])} videos from playlist: {pl_info['title']}")

    playlist_name = pl_info["title"]
    rows = []

    for vid in pl_info["entries"]:
        print(f"Processing video: {vid.get('title', 'Unknown title')}")
        if not vid.get("title"):          # video may be removed/private
            continue

        dur_secs = vid.get("duration") or 0
        duration_minutes = round(dur_secs / 60)   # nearest minute

        rows.append(
            {
                "Playlist name": playlist_name,
                "Title": vid["title"],
                "URL": f"https://www.youtube.com/watch?v={vid['id']}",
                "Duration (minutes)": duration_minutes,
                "Views": vid.get("view_count"),
                "Number of likes": vid.get("like_count"),
                "Number of dislikes": vid.get("dislike_count"),   # may be None
                "Number of comments": vid.get("comment_count"),   # may be None
            }
        )

    return pd.DataFrame(rows)
```

```python
playlists = [
    "https://www.youtube.com/playlist?list=PLz3NrXxHz_CB7AYnBOd-xSUmoCJ3197oM", # MM 3
    "https://www.youtube.com/playlist?list=PLz3NrXxHz_CDMp_ENH22qoEXYv-31tLQJ", # KK 2,
    "https://www.youtube.com/playlist?list=PLz3NrXxHz_CD1urZ5q3bgm_YNWuT2osgS", # MM 4,
    "https://www.youtube.com/playlist?list=PLz3NrXxHz_CBamqkX-nPy-hN2iaHmouyo", # KK 3
    "https://www.youtube.com/playlist?list=PLz3NrXxHz_CA_BlztieZUekYxbVh9XDYG", # KK 4   
]
```

```python
dfs = []

for p in playlists:
    print(f"Processing playlist: {p}")
    df = playlist_to_df(p)
    dfs.append(df)
    df.to_csv(Path("res") / f"{p.split('=')[-1]}.csv", index=False)
```

```python
df_all = pd.concat(dfs, axis=0, ignore_index=True)
df_all.to_csv(Path("res") / "all_playlists.csv", index=False)
```
